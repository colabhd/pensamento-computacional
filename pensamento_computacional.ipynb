##### √çndices e manipula√ß√£o de strings

```python

"""
`[in√≠cio:fim:salto]`
"""
linguagem = 'Python Slice String'
#A fun√ß√£o len() indica o tamanho do tipo de dado. No caso da string, ela indica a quantidade de caracteres
print(len(linguagem))
#O print abaixo j√° mostra o primeiro elemento do conjunto de caracteres
print(linguagem[0])
#O print abaixo ir√° mostrar o √∫ltimo elemento da string
print(linguagem[-1])
#O print abaixo ir√° mostrar o intervalo entre o segundo e o quarto elemento
print(linguagem[2:4])
#O print abaixo ir√° mostrar do primeiro ao quarto elemento
print(linguagem[:4])
#O print abaixo ir√° mostrar do quarto elemento at√© o √∫ltimo
print(linguagem[3:])
#O print abaixo ir√° inverter a string
print(linguagem[::-1])
#O print abaixo segue essa estrutura: `[in√≠cio:fim:salto]`, assim come√ßar√° no √≠ndice 1 e terminar√° no √≠ndice 19, saltando de 2 em 2
print(linguagem[1:19:2])
string_de_numeros = '0123456'
#O print abaixo come√ßa no elemento 1 e vai at√© o √∫ltimo elemento saltando de 2 em 2, passando por todos os √≠ndices √≠mpares
print(string_de_numeros[1::2])
#O print abaixo verifica todos os elementos, saltando de 2 em 2, passando por todos os √≠ndices pares
print(string_de_numeros[::2])
```

##### M√©todos de String üß©

```python
from keyword import iskeyword
#m√©todo split

novo_membro = input('digite o seu nome: ')
print(f'ol√°, {novo_membro}'.split(' '))


#m√©todo join
europa = ['portugal', 'espanha', 'italia']
print(type(europa))
europa_2 = ', '.join(europa)
print(europa_2)
print(type(europa_2))

#m√©todo strip
frase = '     roubei uma goiaba       '
print(frase.strip())

#m√©todos swapcase, upper, lower e title
frase = "Zen do Python"
print(frase.swapcase(), frase.upper(), frase.title(), frase.lower())

#m√©todo replace
print(frase.replace('Zen', 'Bravo'))

#m√©todo count
print(frase.count('n'), type(frase.count('n')))

#m√©todos startswith, endswith, find, isalnum, isalpha, isidentifier, iskeyword, isprintable, isspace, islower e isupper
frase = "Zen do Python"
print(frase.startswith("Zen"))
print(frase.endswith("do"))
print(frase.find('en do Python'))
print(frase.isalnum())
print(frase.isalpha())
print(frase.isidentifier())
print(iskeyword('return'))
print('x\ny'.isprintable())
print(' '.isspace())
print(frase.islower())
print(frase.isupper())

#m√©todo zfill
p = 'palavra'
print(p.zfill(8))

#m√©todos index e find
print(p.index('pa'))
print(p.find('pa'))
```
